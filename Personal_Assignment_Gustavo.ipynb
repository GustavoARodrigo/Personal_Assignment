```python
def create_year(x):
    return int(x.split("-")[0])

def create_month(x):
    return int(x.split("-")[1])

def create_day(x):
    return x[8:10]


db["Year"] = db["Datum"].apply(create_year)
db["Month"] = db["Datum"].apply(create_month)
db["Day"] = db["Datum"].apply(create_day)
```

```python
cols=["Year", "Month"]
db['YM'] = db[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
```

```python
db['Datum'] = pd.to_datetime(db.Datum)
```

##### As we can see, not all measuring stations measure the same gases, so we will analyze NO2, O3, PM2.5 even though PM2.5 is only measured at three of the four stations

The choice was made because these are the gases measured by all stations except PM2.5 and they are the most harmful to the environment, so it makes sense to measure these gases and check if they are at dangerous levels and if so, present measures in order to reduce the emission of the gases or a way to reduce their negative effects.

For O3 and NO2 the thresholds limits are the same, regardless of whether they are evaluated on a 1-hour or 24-hour time frame.

But this is not the case for PM2.5, so we will use the thresholds limits for 24 hours and therefore i decided to do daily resample 

**Threshols limits for each of the gases to be analyzed (24 hours average)**
* NO2:
    * Ideal: NO2 < 50
    * Normal: 50 < NO2 < 100
    * High: 100 < NO2
    
* O3:
    * Ideal: O3 < 65
    * Normal: 65 < O3 < 120
    * High: 120 < O3
    
* PM2.5:
    * Ideal: PM2.5 < 10
    * Normal: 10 < PM2.5 < 20
    * High: 20 < PM2.5

#### NO2

```python
db_rNO2 = db_Rosen[db_Rosen["Parameter"]== "NO2"]
db_rNO2
```

```python
db_rNO2d = db_rNO2.set_index('DATE').resample('D').mean()
db_rNO2d
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
threshold = 0
threshold2 = 50
threshold3 = 100
#threshold4 = 100
x = db_rNO2d.index
for nm in ['Wert']:
    ax.plot(db_rNO2d[nm], label="NO2 Value")
    plt.fill_between(x ,threshold, threshold2, color='green',alpha=0.3)
    plt.fill_between(x ,threshold2, threshold3, color='yellow',alpha=0.3)
    #plt.fill_between(x ,threshold3, threshold4, color='red',alpha=0.3)
    ax.set_ylabel('Values')
    ax.set_title('Zch_Rosengartenstrasse NO2 Values',fontsize=20)
```
